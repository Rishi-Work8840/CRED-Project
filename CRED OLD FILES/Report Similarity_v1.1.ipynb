```python
dff = df[["Report Name","KPI/Used Columns"]].dropna().reset_index()
```

```python
dff.drop(["index"],axis=1,inplace=True)
```

```python
matrix = pd.get_dummies(dff['KPI/Used Columns'])
```

```python
matrix.head()
```

```python
dff = pd.concat([dff, matrix], axis=1)
```

```python
grp = dff.groupby('Report Name').sum()
```

```python
dist = pdist(grp, metric="jaccard")
s_dist = squareform(dist)
# Fill diagonals with nulls
np.fill_diagonal(s_dist, np.nan)
```

```python
sim = np.subtract(1, s_dist)
```

```python
sim_df = pd.DataFrame(sim, columns=grp.index, index=grp.index)
```

```python
sim_df.head()
```

```python
a=[]
for i,r in sim_df.iterrows():
    for j in r.index:
        if r[j] > 0.85:
            # print('1st Report:',i)
            # print('2nd Report:',j)
            # print('Matching %:',round(r[j]*100,3),'\n')
            a.append([i,j,r[j]])
```

```python
len(a)
```

```python
a=[]
print('Report Name','##','No. of matching reports','##','Sum of %','##','Avg Matching %')
for i,r in sim_df.iterrows():
    # print('Report:',i)
    s=0
    x=0
    for j in r.index:
        if r[j]>0.85:
            x+=1
            s+=r[j]
    if x!=0:
        b=[i,x,s,(s*100)/x]
        a.append(b)
        # print(b)
```

```python
print('Report Name','##','No. of matching reports','##','Sum of %','##','Avg Matching %')
for i in a:
    if i[1]>25:
        print(i)
```
