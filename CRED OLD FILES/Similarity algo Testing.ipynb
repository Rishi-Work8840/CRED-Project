```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import squareform, pdist

df =  pd.read_excel("C:/Users/v-yaagarwal/Downloads/KMeans_Data/Test Data.xlsx",dtype="str")
```

```python
df = df[["Report Name","KPI/Used Columns"]].dropna().reset_index(drop=True)
```

```python
df.head()
```

# 100% match
### Report1 = "02 Call Drivers by Module"
### Report2 = "03 Call Drivers by Module First Year Clients"

```python
x = []
for i in df.index:
    if df["Report Name"][i] == "02 Call Drivers by Module":
        x.append(df["KPI/Used Columns"][i])
print("Unique KPI/Columns in Report1:",len(x))
```

```python
y = []
for i in df.index:
    if df["Report Name"][i] == "03 Call Drivers by Module First Year Clients":
        y.append(df["KPI/Used Columns"][i])
print("Unique KPI/Columns in Report2:",len(y))
```

```python
intersection = 0
for i in x:
    if i in y:
        intersection+=1
print("Common columns between them:",intersection)
```

```python
union = len(set(x+y))
print("Total unique columns:",union)
```

```python
Jaccard = (intersection/union)*100
print("Jaccard",round(Jaccard,2),"%")
```
