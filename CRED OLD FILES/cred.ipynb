```python
import pandas as pd
import numpy as np
dataset = pd.read_excel(r"C:\Users\ASHWIK.ACHARYA\Brillio\Mohana Datla - CRED\K-Means\Test Data.xlsx")
```

```python
# 'dataset' holds the input data for this script

from scipy.spatial.distance import squareform, pdist
dataset = dataset[["Domain","Report Name","KPI/Used Columns"]].dropna().reset_index()
dataset.drop(["index"],axis=1,inplace=True)
Total_dataset = pd.DataFrame()
#Running through each domain
for i in dataset.Domain.unique():
    Domain_Dataset = dataset[dataset['Domain'] == i]
    Domain_Dataset = Domain_Dataset[["Report Name","KPI/Used Columns"]].dropna().reset_index()
    Domain_Dataset.drop(["index"],axis=1,inplace=True)
	# matrix = pd.get_dummies(dataset['KPI/Used Columns'],sparse = True ,dtype=int)
    Domain_Dataset = pd.concat([Domain_Dataset, pd.get_dummies(Domain_Dataset['KPI/Used Columns'],sparse = True ,dtype=int)], axis=1).groupby('Report Name').sum()
	#grp = dataset.groupby('Report Name').sum()
    Domain_Dataset.drop('KPI/Used Columns',axis=1,inplace=True)
    dist = pdist(Domain_Dataset, metric="dice")
    s_dist = squareform(dist)
	# Fill diagonals with nulls
    np.fill_diagonal(s_dist, np.nan)
    sim = np.subtract(1, s_dist)
    Domain_Dataset = pd.DataFrame(sim, columns=Domain_Dataset.index, index=Domain_Dataset.index)
    Domain_Dataset = Domain_Dataset.where(np.triu(np.ones(Domain_Dataset.shape)).astype(np.bool_))
    Domain_Dataset.reset_index(inplace = True)
    display(Domain_Dataset)
    break
	#Unpivot to 3 columns
    Domain_Dataset = pd.melt(Domain_Dataset, id_vars = 'Report Name') 
    #Adding Domain name
    Domain_Dataset["Domain"] = i
    Total_dataset = pd.concat([Total_dataset,Domain_Dataset],ignore_index = True)
```
